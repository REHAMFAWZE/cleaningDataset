```python
# Define the correct age ranges for each education level
education_age_ranges = {
    'High School': (15, 19),
    'Associates': (20, 23),  # Ends before Bachelors
    'Bachelors': (24, 27),   # Starts after Associates
    'Masters': (28, 35),     # Starts after Bachelors
    'PhD': (36, 50)          # Starts after Masters
}


# Apply the age range adjustments
for edu, (min_age, max_age) in education_age_ranges.items():
    data.loc[data['Q12 - Highest Level of Education'] == edu, 'Q10 - Current Age'] = \
        data.loc[data['Q12 - Highest Level of Education'] == edu, 'Q10 - Current Age'].clip(min_age, max_age)

```

```python
# Check the min and max age for each education level
age_ranges_updated = data.groupby('Q12 - Highest Level of Education')['Q10 - Current Age'].agg(['min', 'max'])

# Display the updated age ranges
print("Updated Age Ranges After Adjustment:")
print(age_ranges_updated)
```

```python

def assign(age):
  if age==19:
    return 'High School'
  elif 22<=age<=23:
    return 'Associates'
  elif 24<=age<=27:
    return 'Bachelors'
  elif 28<=age<=35:
    return 'Masters'
  else:
    return 'Phd'
data['Q12 - Highest Level of Education'].fillna(data['Q10 - Current Age'].apply(assign),inplace=True)
```
